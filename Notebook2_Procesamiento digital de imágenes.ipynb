```python
# Cargar la imagen
img = cv2.imread('/Users/roxanarachel/Desktop/arbol.jpeg')

# Convertir la imagen a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar umbral binario
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Encontrar contornos
contornos, jerarquia = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujar contornos en la imagen original
img_contornos = img.copy()
cv2.drawContours(img_contornos, contornos, -1, (0, 255, 0), 2)

# Mostrar la imagen original con contornos utilizando matplotlib
plt.imshow(cv2.cvtColor(img_contornos, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Cargar la imagen
img = cv2.imread('/Users/roxanarachel/Desktop/arbol.jpeg')

# Cambiar el tama√±o de la imagen
resized_img = cv2.resize(img, (500, 500))

# Gira la imagen
(filas, columnas) = img.shape[:2]
M = cv2.getRotationMatrix2D((columnas/2, filas/2), 45, 1)
rotated_img = cv2.warpAffine(img, M, (columnas, filas))

# Recortar la imagen
imagen_recortada = img[100:300, 200:400]

# Mostrar las im√°genes utilizando matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(1, 4, 2), plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)), plt.title('Cambio de Tama√±o')
plt.subplot(1, 4, 3), plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)), plt.title('Rotaci√≥n')
plt.subplot(1, 4, 4), plt.imshow(cv2.cvtColor(imagen_recortada, cv2.COLOR_BGR2RGB)), plt.title('Recorte')

plt.show()
```
